## Project Overview

This notebook analyzes loan approval decisions using a structured business analytics approach. The goal is to identify which factors most strongly influence approval outcomes and create a credit-based risk framework that improves decision consistency, fairness, and operational efficiency.

The original report analyzed 4,269 loan applications and found that credit score was the dominant driver of approval outcomes.


## Business Question

How can financial institutions use data to make loan approval decisions more consistent, fair, and risk-aware?

This notebook focuses on:

- Approval rate patterns
- Credit score segmentation
- Loan term impact
- Feature importance
- Risk-tier decision framework


````python
# Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", None)
pd.set_option("display.width", 1000)
````

## Step 1: Create Sample Loan Approval Dataset

This sample dataset is built to reflect the findings from the final project report. It includes borrower information such as credit score, income, loan amount, loan term, education, employment type, and approval outcome.

For a real GitHub portfolio project, this notebook can later be connected to the original Kaggle dataset.


````python
# Set random seed for reproducibility
np.random.seed(42)

# Create sample dataset size based on final report
n = 4269

# Generate credit scores
credit_score = np.random.randint(300, 901, n)

# Generate other loan application variables
income = np.random.randint(25000, 180000, n)
loan_amount = np.random.randint(5000, 350000, n)
loan_term = np.random.choice([2, 4, 6, 8, 10, 12], n)
education = np.random.choice(["Graduate", "Not Graduate"], n)
self_employed = np.random.choice(["Yes", "No"], n)

# Approval logic based mostly on credit score
approval_probability = np.where(
    credit_score >= 750, 0.99,
    np.where(credit_score >= 650, 0.75,
    np.where(credit_score >= 600, 0.45,
    np.where(credit_score >= 500, 0.25, 0.10)))
)

# Slight loan term adjustment
approval_probability = approval_probability - (loan_term * 0.01)
approval_probability = np.clip(approval_probability, 0.02, 0.99)

loan_status = np.where(np.random.rand(n) < approval_probability, "Approved", "Rejected")

# Create DataFrame
df = pd.DataFrame({
    "credit_score": credit_score,
    "income": income,
    "loan_amount": loan_amount,
    "loan_term_years": loan_term,
    "education": education,
    "self_employed": self_employed,
    "loan_status": loan_status
})

df.head()
````

## Step 2: Dataset Overview

This section checks the dataset structure, missing values, and overall approval rate.


````python
# Dataset shape
print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")

# Missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Approval rate
approval_rate = (df["loan_status"] == "Approved").mean() * 100
rejection_rate = (df["loan_status"] == "Rejected").mean() * 100

print(f"\nApproval Rate: {approval_rate:.1f}%")
print(f"Rejection Rate: {rejection_rate:.1f}%")
````

````python
# Summary statistics
df.describe()
````

## Step 3: Approval Status Distribution

This chart shows the overall split between approved and rejected loan applications.


````python
# Approval distribution

status_counts = df["loan_status"].value_counts()

plt.figure(figsize=(7, 5))
plt.bar(status_counts.index, status_counts.values)
plt.title("Loan Approval Distribution")
plt.xlabel("Loan Status")
plt.ylabel("Number of Applications")
plt.grid(axis="y")
plt.show()
````

## Step 4: Credit Score Segmentation

Credit score was the strongest predictor in the final report. This section groups applicants into credit score bands and calculates approval rates by segment.


````python
# Create credit score bands

bins = [299, 500, 600, 650, 750, 900]
labels = ["300-500", "501-600", "601-650", "651-750", "751-900"]

df["credit_score_band"] = pd.cut(df["credit_score"], bins=bins, labels=labels)

# Approval rate by credit score band
credit_band_summary = df.groupby("credit_score_band")["loan_status"].apply(
    lambda x: (x == "Approved").mean() * 100
).reset_index(name="approval_rate")

credit_band_summary
````

````python
# Plot approval rate by credit score band

plt.figure(figsize=(9, 5))
plt.bar(credit_band_summary["credit_score_band"].astype(str), credit_band_summary["approval_rate"])
plt.title("Approval Rate by Credit Score Band")
plt.xlabel("Credit Score Band")
plt.ylabel("Approval Rate (%)")
plt.ylim(0, 100)
plt.grid(axis="y")
plt.show()
````

## Step 5: Loan Term and Risk

The final report found that loan term influences approval decisions. Shorter repayment periods tend to reduce perceived risk.


````python
# Approval rate by loan term

term_summary = df.groupby("loan_term_years")["loan_status"].apply(
    lambda x: (x == "Approved").mean() * 100
).reset_index(name="approval_rate")

term_summary
````

````python
# Plot approval rate by loan term

plt.figure(figsize=(8, 5))
plt.plot(term_summary["loan_term_years"], term_summary["approval_rate"], marker="o")
plt.title("Approval Rate by Loan Term")
plt.xlabel("Loan Term (Years)")
plt.ylabel("Approval Rate (%)")
plt.grid(True)
plt.show()
````

## Step 6: Demographic Variables

The report found that education and self-employment status had minimal impact on approval decisions. This section compares approval rates across those variables.


````python
# Approval rate by education

education_summary = df.groupby("education")["loan_status"].apply(
    lambda x: (x == "Approved").mean() * 100
).reset_index(name="approval_rate")

education_summary
````

````python
# Approval rate by self-employment status

employment_summary = df.groupby("self_employed")["loan_status"].apply(
    lambda x: (x == "Approved").mean() * 100
).reset_index(name="approval_rate")

employment_summary
````

## Step 7: Feature Importance Summary

The final report found that credit score had the highest predictive importance, followed by loan term. Other variables had minimal explanatory value.

This section recreates the feature importance summary as a business-friendly visualization.


````python
# Feature importance based on final report findings

feature_importance = pd.DataFrame({
    "Feature": ["Credit Score", "Loan Term", "Income", "Loan Amount", "Education", "Self-Employed"],
    "Importance": [0.803, 0.118, 0.019, 0.018, 0.012, 0.010]
})

feature_importance
````

````python
# Plot feature importance

feature_importance_sorted = feature_importance.sort_values("Importance", ascending=True)

plt.figure(figsize=(9, 5))
plt.barh(feature_importance_sorted["Feature"], feature_importance_sorted["Importance"])
plt.title("Feature Importance for Loan Approval Decisions")
plt.xlabel("Importance Score")
plt.ylabel("Feature")
plt.grid(axis="x")
plt.show()
````

## Step 8: Credit-Based Risk Tier Framework

Based on the analysis, the proposed framework uses credit score as the main decision factor and routes applicants into approval tiers.


````python
# Define risk tier function

def assign_risk_tier(score):
    if score >= 750:
        return "High Tier - Automatic Approval"
    elif score >= 650:
        return "Medium Tier - Conditional Approval"
    elif score >= 600:
        return "Borderline Tier - Manual Review"
    else:
        return "Low Tier - Reject"

df["risk_tier_decision"] = df["credit_score"].apply(assign_risk_tier)

df[["credit_score", "loan_status", "risk_tier_decision"]].head(10)
````

````python
# Count applicants by risk tier

risk_tier_counts = df["risk_tier_decision"].value_counts().reset_index()
risk_tier_counts.columns = ["Risk Tier Decision", "Application Count"]

risk_tier_counts
````

````python
# Visualize risk tier distribution

plt.figure(figsize=(10, 5))
plt.bar(risk_tier_counts["Risk Tier Decision"], risk_tier_counts["Application Count"])
plt.title("Applications by Risk Tier Decision")
plt.xlabel("Risk Tier")
plt.ylabel("Number of Applications")
plt.xticks(rotation=30, ha="right")
plt.grid(axis="y")
plt.tight_layout()
plt.show()
````

## Step 9: Business Recommendations

Based on the analysis, the recommended loan approval framework includes:

1. **Automatic approval** for applicants with credit scores of 750+
2. **Conditional approval** for applicants between 650–749
3. **Manual review** for borderline applicants between 600–649
4. **Rejection or additional documentation** for applicants below 600
5. Reduced reliance on low-impact variables such as education and self-employment status

This approach improves decision consistency while reducing unnecessary bias.


## Final Insights

The analysis supports a credit-based risk segmentation framework for loan approval decisions.

### Key Takeaways

- Credit score is the strongest driver of loan approval outcomes.
- Loan term also influences perceived lending risk.
- Education and self-employment status showed minimal impact.
- A structured approval framework can improve fairness, consistency, and efficiency.
